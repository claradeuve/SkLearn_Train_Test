```python
df_flights = df_flights.fillna(0)
count_nan_in_df = df_flights.isnull().sum()
print (count_nan_in_df)
```

```python
from sklearn.model_selection import train_test_split

# Separar dataset en features y labels
y = df_flights.ArrDelay
X = df_flights.drop('ArrDelay', axis=1)

# Dividir el dataset en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

## Exercici 2
Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...)

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

# Separar dataset en atributos numéricos y categóricos
X_int = X.loc[:,(X.dtypes=='int').values]
X_float = X.loc[:,(X.dtypes=='float').values]
X_num = pd.concat([X_int, X_float], axis=1)
display(X_num)


# Generar dummies de atributos categóricos
X_ = X.loc[:,(X.dtypes=='object').values]
X_dummy = pd.get_dummies(X_)
display(X_dummy)


# Generar polinomios de atributos numéricos
poly = PolynomialFeatures(2)
X_poly = poly.fit_transform(X_num)
X_poly = pd.DataFrame(X_poly)
display(X_poly)


# Estandarizar los polinomios de atributos numéricos
scaler = StandardScaler()
X_poly_scaled = scaler.fit_transform(X_poly)
X_poly_scaled = pd.DataFrame(X_poly_scaled)
display(X_poly_scaled)
```

He decidido generar polinomios para crear interacciones entre variables y estandarizarlos después para aumentar su señal.

He optado por no estandarizar primero porque, al estandarizar, los datos se convierten en valores z, que generalmente están entre -3 y 3. Por tanto, si hubiera estandarizado antes, con los polinomios hubiera obtenido múltiplos y cuadrados de estos valores entre -3 y 3, y, por consiguiente, valores de una magnitud más pequeña que la original.

Asimismo, estandarizando después se evita crear valores negativos al azar. Al estandarizar, el conjunto de valores positivos se convierte en un conjunto de valores tanto positivos como negativos (con la media en cero). Estos nuevos valores estandarizados negativos generarían más valores negativos si se decide crear los polinomios a partir de ellos, ya que se multiplicarían entre sí.
